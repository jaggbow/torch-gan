```python
import sys
import os
import torch
from torch import dtype
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path+"\\GAN")
from gan import *
import seaborn as sns
import matplotlib.pyplot as plt

dtype = float
device = torch.device("cpu")
```

```python
#Dataset
N=10000
norm = np.random.normal(0, 0.5, N)
x=np.linspace(-50,50,N)
sns.scatterplot(x,x**2+norm)
```

```python
latent_dim = 2
advNet = GAN(latent_dim,device,discriminator_lr=0.001, generator_lr=0.001)
```

```python
X = torch.from_numpy(np.stack([x,x**2+norm]).T).float()
```

```python
advNet.train(X, batch_size=128, epochs=500)
```

```python
plt.plot(advNet.history['Generative'])
```

```python
plt.plot(advNet.history['Discriminator'])
```

```python
pred = advNet.generate(10000)
```

```python
pred = pred.cpu().detach().numpy()
```
